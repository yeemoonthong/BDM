### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

```python
pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep="\t")
```

### Step 3. Assign it to a variable called chipo.

```python
chip = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep="\t")

# Remove the $ sign and convert the item_price field to a float
chip['item_price'] = chip['item_price'].str.replace('$', '').astype(float)
```

### Step 4. How many products cost more than $10.00?

```python
len(chip.query("item_price > 10"))
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
# Select all the rows and only the columns of item_name and item_price
chip.loc[:, ['item_name', 'item_price']].groupby('item_name').min().reset_index()
```

### Step 6. Sort by the name of the item

```python
chip.sort_values(by="item_name")
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
# Get the index of the item with the max item_price
item_index = chip["item_price"].idxmax()

chip.loc[item_index]["quantity"]
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
len(chip.query("item_name == 'Veggie Salad Bowl'"))
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
len(chip.query("item_name == 'Canned Soda' and quantity > 1"))
```
